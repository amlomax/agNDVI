```python
import matplotlib.pyplot as plt
import numpy as np
import rasterio 
import glob
```

```python
# TO DO: put all landsat images into data/landsat
# Define Landsat
cv_2023_files = sorted(glob.glob('data/landsat/cv09102023/*TIF'))
cv_2023_files
```

```python
# NDVI
src_4 = rasterio.open(cv_2023_files[3])
band_4 = src_4.read(1)

src_5 = rasterio.open(cv_2023_files[4])
band_5 = src_5.read(1)
```

```python
np.seterr(divide='ignore', invalid='ignore')
ndvi = (band_5.astype(float) - band_4.astype(float)) / (band_5.astype(float) + band_4.astype(float))
```

```python
# Plot NDVI image 
fig, ax = plt.subplots(figsize=(8,8))
im = ax.imshow(ndvi, cmap='Greens')
ax.set_title("Central Valley NDVI 09/10/2023")
fig.colorbar(im, orientation='horizontal')
plt.show()
```
